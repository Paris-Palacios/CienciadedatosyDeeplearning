# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# 1. Figuras geométricas


### Creación de una imagen en color negro

```python
# Imagen color negro de 512 x 512 
canvas = np.zeros( shape=(512,512,3) , dtype=np.int16 )
canvas.shape
```

```python
plt.imshow(canvas)
plt.show()
```

#### Rectangulo

```python
#cv.rectangle(imagen,origen,final,color,grosor de linea)
cv.rectangle(canvas, pt1=(400,0) ,  pt2=(500,100) , color=(0,255,0) , thickness = 3   );
cv.rectangle(canvas, pt1=(200,200) ,  pt2=(400,400) , color=(255,0,0) , thickness = 3   );
```

```python
plt.imshow(canvas)
plt.show()
```

#### Circulo

```python
#cv.circle(imagen, centro, radio, color, grosor de linea)
cv.circle(canvas, center=(200,200), radius=50, color=(0,255,0) , thickness = 3   ) ;
cv.circle(canvas, center=(200,200), radius=100, color=(0,255,0) , thickness = 3   ) ;
cv.circle(canvas, center=(200,200), radius=20, color=(0,255,0) , thickness = -1   ) ;
```

```python
plt.imshow(canvas)
plt.show()
```

#### Linea

```python
#cv.line(imagen, inicio, fin, color, grosor de linea)
cv.line(canvas,  pt1=(400,0) ,  pt2=(500,100), color=(0,255,0) , thickness = 2 );
cv.line(canvas,  pt1=(0,0) ,  pt2=(512,512), color=(0,255,0) , thickness = 2 );
```

```python
plt.imshow(canvas)
plt.show()
```

#### Texto

```python
#font = cv.FONT_HERSHEY_SIMPLEX
#lineType=cv.LINE_AA
#cv.putText(imagen,texto,origen, fuente,tamaño de la fuente,color,grosor, tipo de linea)
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(canvas,text='HOLA',org=(10,400) ,
           fontFace=cv.FONT_HERSHEY_SIMPLEX,fontScale=3,
           color=(0,255,0),thickness=2, lineType=cv.LINE_AA);
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(canvas,text='HOLA',org=(10,500) ,
           fontFace=cv.FONT_HERSHEY_SIMPLEX,fontScale=2,
           color=(0,255,255),thickness=2, lineType=cv.LINE_AA);
```

```python
plt.imshow(canvas)
plt.show()
```


#### Polígono 

```python
# Imagen color negro de 512 x 512 
canvas2 = np.zeros( shape=(512,512,3) , dtype=np.int16 )
plt.imshow(canvas2)
plt.show()
```

```python
#cv.polylines(imagen,puntos,Cerrada o abierta,color y grosor)
puntos = np.array([ [100,200], [400,200], [400,500]  ] )
cv.polylines(canvas2, [puntos], isClosed=True , color=(0,255,0) , thickness = 2 );
```

# 2. Rotación de imagenes

## 2.1 Función *flip*

```python
# archivo figuras/wave.png
img=cv.imread('figuras/wave.png')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
new_img = cv.flip(img, 1)
plt.imshow(new_img)
plt.show()
```

```python
new_img = cv.flip(img, 0)
plt.imshow(new_img)
plt.show()
```

```python
new_img = cv.flip(img, -1)
plt.imshow(new_img)
plt.show()
```

## 3. Detección de bordes

```python
# figuras/coins.png
img = cv.imread('figuras/coins.png')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img.shape
```

```python
img_gray=cv.cvtColor(img,cv.COLOR_RGB2GRAY)
plt.imshow(img_gray,cmap='gray')
plt.show()
```

```python
img_gauss=cv.GaussianBlur(img_gray,(3,3,),0)
plt.imshow(img_guass,cmap='gray')
plt.show()
```

```python
thr, img_thr = cv.threshold (img_gauss, 160,255, cv.THRESH_BINARY ) 
plt.imshow(img_thr,cmap='gray')
plt.show()
```

```python
thr, img_thr = cv.threshold (img_gauss, 155,255, cv.THRESH_BINARY_INV ) 
plt.imshow(img_thr,cmap='gray')
plt.show()
```

```python
contornos, _ =   cv.findContours(img_thr, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE )
```

```python
# figuras/coins.png
img = cv.imread('figuras/coins.png')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

cv.drawContours( img, contornos , -1 , (0,255,0), 2   )

plt.imshow(img)
plt.show()
```

```python
# figuras/coins.png
img = cv.imread('figuras/coins.png')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

cv.drawContours( img, contornos , 1 , (0,255,0), 2   )

plt.imshow(img)
plt.show()
```
