```python
img = cv.imread('figuras/monedas.jpeg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_gray=cv.cvtColor(img,cv.COLOR_RGB2GRAY)
img_gauss=cv.GaussianBlur(img_gray,(3,3,),0)
thr, img_thr = cv.threshold (img_gauss, 170,255, cv.THRESH_BINARY_INV ) 
contornos, _ =   cv.findContours(img_thr, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE )
cv.drawContours( img, contornos , -1 , (255,0,0), 2   )

plt.imshow(img)
plt.show()
```
