# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
def plt_imshow(title, image):
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

# 1. Operaciones lógicas


 ## Implementación de operadores OpenCV AND, OR, XOR y NOT  

<div class="alert alert-info">

<img align="left" width="500"  float= "none" align="middle" src="figuras/rectangulo_circulo.png">

```python
# Rectangulo 

rectangle = np.zeros((300, 300), dtype="uint8")
cv.rectangle(rectangle, (25, 25), (275, 275), 255, -1)
plt_imshow("Rect", rectangle)

```

```python
# Circulo

circle = np.zeros((300, 300), dtype = "uint8")
cv.circle(circle, (150, 150), 150, 255, -1)
plt_imshow("Circulo", circle)
```

<div class="alert alert-info">
 <ul>
  <li>Aplicar operaciones de AND, OR y XOR y NOT</li>
</ul> 


```python
#AND
img_and = cv.bitwise_and(rectangle,circle)
plt_imshow('AND',img_and)
```

```python
#OR
img_or = cv.bitwise_or(rectangle,circle)
plt_imshow('OR',img_or)
```

```python
#XOR
img_xor = cv.bitwise_xor(rectangle,circle)
plt_imshow('XOR',img_xor)
```

<div class="alert alert-info">
 <ul>
  <li>Aplicar la operacón inversa a ambas imagenes</li>
</ul> 

```python
#Rectangulo
img_not = cv.bitwise_not(rectangle)
plt_imshow('NOT',img_not)
```

```python
#Circulo
img_not = cv.bitwise_not(circle)
plt_imshow('NOT',img_not)
```

# 2. Máscaras

```python
# Abrir el archivo 'figuras/people.jpg'
img  = cv.imread('figuras/people.jpg')
plt_imshow('original',img)
```

```python
# Mascara rectangular

mascara_rectangulo = np.zeros(img.shape[:2], dtype="uint8")
cv.rectangle(mascara_rectangulo, (90, 50), (230, 250), 255, -1)
plt_imshow("Mascara rectangulo", mascara_rectangulo)
```

```python
#Aplicar la mascara rectangular
mascara = cv.bitwise_and(img,img,mask = mascara_rectangulo)
plt_imshow("Mascara aplicada", mascara)
```

```python
# Mascara circular y su implementación

mascara_circulo = np.zeros(img.shape[:2], dtype = "uint8")
cv.circle(mascara_circulo, (150, 130), 100, 255, -1)
plt_imshow("Circulo", mascara_circulo)

mascara = cv.bitwise_and(img,img,mask = mascara_circulo)
plt_imshow("Mascara aplicada", mascara)
```

# 3. Operaciones morfológicas 

```python
# Imagen figuras/labdatos.png en escala de grises
img = cv.imread('figuras/labdatos.png')
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.figure(figsize=(10,6))
plt_imshow('Escala de grises',gray)
```

### Erosión

```python
#Erosión
erosion = cv.erode(gray.copy(),None,iterations=1)
plt.figure(figsize=(10,6))
plt_imshow('Erosion, i=1 ',erosion)
```

```python
# Varias erosiones
erosion = cv.erode(gray.copy(),None,iterations=2)
plt.figure(figsize=(10,6))
plt_imshow('Erosion, i=2 ',erosion)

erosion = cv.erode(gray.copy(),None,iterations=3)
plt.figure(figsize=(10,6))
plt_imshow('Erosion, i=3 ',erosion)
```

### Dilatación

```python
#Dilatación
dilatacion = cv.dilate(gray.copy(),None,iterations=1)
plt.figure(figsize=(10,6))
plt_imshow('Dilatacion, i=1 ',dilatacion)
```

```python
#La dilatación varias veces
dilatacion = cv.dilate(gray.copy(),None,iterations=2)
plt.figure(figsize=(10,6))
plt_imshow('Dilatacion, i=2 ',dilatacion)

dilatacion = cv.dilate(gray.copy(),None,iterations=3)
plt.figure(figsize=(10,6))
plt_imshow('Dilatacion, i=3 ',dilatacion)
```
